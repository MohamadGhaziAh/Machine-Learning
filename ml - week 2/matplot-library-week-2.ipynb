```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Data
x = np.arange(10,20,0.1)
y =np.tan(x)
plt.plot(x,y)
plt.title("Tan Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
x = np.arange(10,20,0.1)
y =np.cos(x)
plt.plot(x,y)
plt.title("cosine Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
x = np.random.rand(100)
y= np.random.rand(100)
colors = np.random.rand(100)
sizes = np.random.randint(100,200,100)

plt.scatter(x,y, c=colors, s=sizes)
plt.title("Pint Chart")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show
```

```python
data = np.random.randn(200)
plt.hist(data, bins=23)
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
x = ['M','O','H','A','M','A','M']
y = np.random.randint(1,10,7)
plt.bar(x,y)
plt.title("Chart")
plt.xlabel("LETS")
plt.ylabel("NUMBERS")
plt.show()
```

```python
sizes = [80,10,5,2,2,1]
plt.pie(sizes)
plt.title("diskspace")
plt.show()
```
